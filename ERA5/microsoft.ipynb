```python
from utils import WindSpeedVisualizer, PlanetaryComputerERA5
import pandas as pd
```

```python
pc = PlanetaryComputerERA5('https://planetarycomputer.microsoft.com/api/stac/v1/')
date_range = pd.date_range("2020-01-01", periods=24, freq="h")
variables=["northward_wind_at_10_metres","eastward_wind_at_10_metres"]
pc.get_data(date_range=date_range, variables=variables)
pc.download()
```

```python
wind_speed_pc, _ = pc.calculate_wind_speed()
```

```python
WindSpeedVisualizer.plot_wind_speed(wind_speed_pc.sel(time=wind_speed_pc.time[0]))
```

```python
wind_anim = WindSpeedVisualizer.generate_animation(wind_speed_pc)
```
