```python
from utils import GridInterpolator
from utils import ProductdVisualizer
from utils import load_config, Cmems
```

```python
config = load_config(file_path="config.yaml")
```

```python
c = Cmems(config["cmems_request_bgc"])
bgc_path = c.download_data()
c = Cmems(config["cmems_request_glo_phy"])
glo_path = c.download_data()
gi = GridInterpolator(dataset_input_path=glo_path, dataset_target_path=bgc_path, variable="thetao")
ds_glo, ds_bgc  = gi.interpolate()
```

```python
param = {"log_norm": False,
         "cmap": "coolwarm",
         "title": "Sea water potential temperature [°C]",
         "unit": "°C"}

pv = ProductdVisualizer(param=param, ds=ds_glo.thetao)
anim = pv.generate_animation()
anim
```

```python
param = {"log_norm": True,
         "cmap": "Greens",
         "title": "Mass concentration of chlorophyll a in sea water [mg/m3]",
         "unit": "mchl [mg/m3]"}

pv = ProductdVisualizer(param=param, ds=ds_bgc.chl)
anim = pv.generate_animation()
anim
```
